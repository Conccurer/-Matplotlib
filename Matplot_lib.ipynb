## Plot Customization:

```python
years = [2014 + x for x in range(8)]
income = [55, 56, 62, 61, 72, 72, 73, 75]
income_ticks = list(range(50, 81, 2))
```

```python
plt.plot(years, income)
plt.title("Income per Years",fontsize=20,fontname='cursive')
plt.xlabel("Years")
plt.ylabel("Income (thousand USD)")
plt.yticks(income_ticks, [f'{x}K USD' for x in income_ticks] )
plt.show()
```

## Subplots:

```python
x = np.arange(100)
fig, axs = plt.subplots(2,2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cos Wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log(range(1,101)))
axs[1, 1].set_title("log Wave")
axs[1, 1].set_xlabel("X")

plt.suptitle("Four Plots")
plt.tight_layout()
# plt.savefig("Four_plt.png", dpi=300, transparent=True, bbox_inches='tight', pad_inches=2)   # Save the image of this graph
```

## 3-D Plotting:

```python
ax = plt.axes(projection='3d')
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3D Plot")
plt.show()
```

```python
ax = plt.axes(projection='3d')
x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)
ax.plot(x, y, z)
ax.set_title("3D Line Plot")
plt.show()
```

```python
ax = plt.axes(projection='3d')

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)
ax.plot_surface(X, Y, Z)
ax.set_title("3D Surface Plot")
plt.show()
```

## Animation Plot:

```python
import random
```

```python
heads_Tails = [0, 0]

for _ in range(100000):
    heads_Tails[random.randint(0, 1)] += 1
    plt.bar(['Heads', "Tails"], heads_Tails, color=['red', 'blue'])
    plt.pause(0.001)
plt.show()
```
